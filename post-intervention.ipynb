### Frequency of post intervention data

```python
def calculate_digital_tool_usage(transformed_data, column_name, mapping_dict, main_column_name):
    """
    Function to calculate the usage statistics of digital tools based on a provided data column and mapping.
    
    Parameters:
    - transformed_data: DataFrame containing the transformed data.
    - column_name: The column name for digital tool usage frequency.
    - mapping_dict: Dictionary mapping the actual usage categories to their indices.
    
    Returns:
    - DataFrame with Digital Tool Usage, Count, and Percentage.
    - Total count and total percentage of digital tool usage.
    """
    # Get the index and values from value_counts
    usage_index = transformed_data[column_name].value_counts().index
    usage_value = transformed_data[column_name].value_counts().values
    
    # Map the index to the actual values using the provided dictionary
    usage_index_mapped = [list(mapping_dict.keys())[list(mapping_dict.values()).index(i)] for i in usage_index]
    
    # Create a DataFrame with usage statistics
    data = {main_column_name: usage_index_mapped, 'Count': usage_value}
    df = pd.DataFrame(data)
    
    # Calculate total count and percentage
    total_count = df['Count'].sum()
    df['Percentage'] = ((df['Count'] / total_count) * 100).round(1)
    
    # Calculate total percentage
    total_percentage = df['Percentage'].sum()
    
    # Print results
    print(f'Total Count: {total_count}')
    print(f'Total Percentage: {total_percentage}')
    
    return df
```

### Perform paired t-tests or Wilcoxon tests for significance.

```python
from scipy.stats import ttest_rel, wilcoxon

# Ensure we only keep rows where both pre and post intervention metrics exist for the relevant columns
aligned_pre_data = pre_intervention_data[['AI_Tool_Usage', 'Frequency_of_Digital_Tool_Usage', 'Comfort_with_Technology']].dropna()
aligned_post_data = post_intervention_data[['improvement_in_ai_familiarity', 'reduction_in_time_spent_create_materials', 'confidence_in_using_ai_in_teaching']].dropna()

# Align the indices to make sure we are comparing the same participants
common_indices = aligned_pre_data.index.intersection(aligned_post_data.index)

# Filter both datasets by the common indices
aligned_pre_data = aligned_pre_data.loc[common_indices]
aligned_post_data = aligned_post_data.loc[common_indices]

# Now perform Paired t-tests
ttest_results_aligned = {
    "AI Familiarity (Pre vs Post)": ttest_rel(aligned_pre_data['AI_Tool_Usage'], aligned_post_data['improvement_in_ai_familiarity']),
    "Time Spent Creating Materials (Pre vs Post)": ttest_rel(aligned_pre_data['AI_Tool_Usage'], aligned_post_data['reduction_in_time_spent_create_materials']),
    "Comfort with Technology (Pre vs Post)": ttest_rel(aligned_pre_data['AI_Tool_Usage'], aligned_post_data['confidence_in_using_ai_in_teaching'])
}

# Perform Wilcoxon Signed-Rank tests for non-parametric data
wilcoxon_results_aligned = {
    "AI Familiarity (Pre vs Post)": wilcoxon(aligned_pre_data['AI_Tool_Usage'], aligned_post_data['improvement_in_ai_familiarity']),
    "Time Spent Creating Materials (Pre vs Post)": wilcoxon(aligned_pre_data['AI_Tool_Usage'], aligned_post_data['reduction_in_time_spent_create_materials']),
    "Comfort with Technology (Pre vs Post)": wilcoxon(aligned_pre_data['AI_Tool_Usage'], aligned_post_data['confidence_in_using_ai_in_teaching'])
}
```

### Calculate effect sizes for practical significance.

```python
import numpy as np

# Function to calculate Cohen's d for effect size
def cohen_d(x, y):
    """Calculate Cohen's d for effect size between two paired samples."""
    n = len(x)
    diff = x - y
    mean_diff = np.mean(diff)
    pooled_std = np.sqrt((np.std(x, ddof=1) ** 2 + np.std(y, ddof=1) ** 2) / 2)
    return mean_diff / pooled_std

# Now calculate effect sizes (Cohen's d) for each of the comparisons

# AI Familiarity (Pre vs Post)
effect_size_ai_familiarity = cohen_d(aligned_pre_data['AI_Tool_Usage'], aligned_post_data['improvement_in_ai_familiarity'])

# Time Spent Creating Materials (Pre vs Post)
effect_size_time_spent = cohen_d(aligned_pre_data['Frequency_of_Digital_Tool_Usage'], aligned_post_data['reduction_in_time_spent_create_materials'])

# Comfort with Technology (Pre vs Post)
effect_size_comfort_tech = cohen_d(aligned_pre_data['Comfort_with_Technology'], aligned_post_data['confidence_in_using_ai_in_teaching'])

# Create a dictionary of the effect sizes
effect_sizes = {
    "AI Familiarity (Pre vs Post)": effect_size_ai_familiarity,
    "Time Spent Creating Materials (Pre vs Post)": effect_size_time_spent,
    "Comfort with Technology (Pre vs Post)": effect_size_comfort_tech
}
effect_sizes_df = pd.DataFrame(list(effect_sizes.items()), columns=["Metric", "Cohen's d (Effect Size)"])
effect_sizes_df
```
